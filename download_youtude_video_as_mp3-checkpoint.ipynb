```python
from __future__ import unicode_literals
import youtube_dl
import time
import requests
################################

response = requests.get(input("ç¶²å€ :"))
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())

net0 = str(soup)
net0 = net0.split('playlist-items')
#net0 = str(input("åŸå§‹ç¢¼è¼¸å…¥ :")) # åªæ“·å– htmlä¸­ playlist çš„éƒ¨åˆ†

seconds1 = time.time()
net = net0.split("\"")

youtube_nets = [] # å·²ç¶“æ˜¯å¯ä»¥æ’­æ”¾çš„youtubeç¶²å€
for n in net:
    if "/watch" in n:
        youtube_nets.append('https://www.youtube.com'+ n) # å·²ç¶“æ˜¯å¯ä»¥æ’­æ”¾çš„youtubeç¶²å€
        
#ç¶²å€ä¸èƒ½åŒ…å« "&list=RDH81DzDMdCLI"
nolist_youtube_nets = [] #æ²’æœ‰ "&list=RDH81DzDMdCLI "é …é€ æˆä¸èƒ½ä¸‹è¼‰
for n in youtube_nets:
    ns = n.split("&")
    n0 = '' # è¢«åŠ ç©ºç¶²å€
    for i,j in enumerate(ns):
        if not "list" in j:
            n0 = n0 + j
            if i > 0:
                n0 = n0 + "&" + j
                
# å·²ç¶“æ˜¯å¯ä»¥æ’­æ”¾çš„youtubeç¶²å€listï¼Œä¸”æ²’æœ‰ "&list=RDH81DzDMdCLI "é …é€ æˆä¸èƒ½ä¸‹è¼‰                
    nolist_youtube_nets.append(n0)

print(nolist_youtube_nets)

for i,j in enumerate(nolist_youtube_nets):
    if i % 2 == 0:
        print(j)
        
        try:
            ydl_opts = {
                'format': 'bestaudio/best',
                'postprocessors': [{
                    'key': 'FFmpegExtractAudio',
                    'preferredcodec': 'mp3',
                    'preferredquality': '192',
                }],
            }
            with youtube_dl.YoutubeDL(ydl_opts) as ydl:
                ydl.download([j])
        except:
            print("download failure\n", j)
        
            
seconds2 = time.time()
print('time', seconds2 - seconds1)
```
