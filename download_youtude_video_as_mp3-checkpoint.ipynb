```python
from __future__ import unicode_literals
import youtube_dl
import time
import requests
################################

response = requests.get(input("Á∂≤ÂùÄ :"))
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())

net0 = str(soup)
net0 = net0.split('playlist-items')
#net0 = str(input("ÂéüÂßãÁ¢ºËº∏ÂÖ• :")) # Âè™Êì∑Âèñ html‰∏≠ playlist ÁöÑÈÉ®ÂàÜ

seconds1 = time.time()
net = net0.split("\"")

youtube_nets = [] # Â∑≤Á∂ìÊòØÂèØ‰ª•Êí≠ÊîæÁöÑyoutubeÁ∂≤ÂùÄ
for n in net:
    if "/watch" in n:
        youtube_nets.append('https://www.youtube.com'+ n) # Â∑≤Á∂ìÊòØÂèØ‰ª•Êí≠ÊîæÁöÑyoutubeÁ∂≤ÂùÄ
        
#Á∂≤ÂùÄ‰∏çËÉΩÂåÖÂê´ "&list=RDH81DzDMdCLI"
nolist_youtube_nets = [] #Ê≤íÊúâ "&list=RDH81DzDMdCLI "È†ÖÈÄ†Êàê‰∏çËÉΩ‰∏ãËºâ
for n in youtube_nets:
    ns = n.split("&")
    n0 = '' # Ë¢´Âä†Á©∫Á∂≤ÂùÄ
    for i,j in enumerate(ns):
        if not "list" in j:
            n0 = n0 + j
            if i > 0:
                n0 = n0 + "&" + j
                
# Â∑≤Á∂ìÊòØÂèØ‰ª•Êí≠ÊîæÁöÑyoutubeÁ∂≤ÂùÄlistÔºå‰∏îÊ≤íÊúâ "&list=RDH81DzDMdCLI "È†ÖÈÄ†Êàê‰∏çËÉΩ‰∏ãËºâ                
    nolist_youtube_nets.append(n0)

print(nolist_youtube_nets)

for i,j in enumerate(nolist_youtube_nets):
    if i % 2 == 0:
        print(j)
        
        try:
            ydl_opts = {
                'format': 'bestaudio/best',
                'postprocessors': [{
                    'key': 'FFmpegExtractAudio',
                    'preferredcodec': 'mp3',
                    'preferredquality': '192',
                }],
            }
            with youtube_dl.YoutubeDL(ydl_opts) as ydl:
                ydl.download([j])
        except:
            print("download failure\n", j)
        
            
seconds2 = time.time()
print('time', seconds2 - seconds1)
```
